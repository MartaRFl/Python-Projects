```python
from random import randint

def check_user_guess(user_guess):
    while (user_guess<1 or user_guess>100):
        print("Your number is not valid, please choose another number.")
        user_guess= int(input("Introduce a number between 1 and 100: "))
    return user_guess

def compare_number(number,user_guess):
    if number == user_guess:
        return True
    else:
        return False

def generate_clue(number):
    clue=randint(1,3)
    comparing=randint(1,100)
    bigger_than= number>comparing
    
    if clue == 1:
        if number > comparing:
            print("The number is bigger than " + str(comparing))
        else:
            print("The number is lower or equal to "+ str(comparing))
            
    elif clue == 2:
        divisible=randint(2,3)
        if number%divisible == 0:
            print("The number divisible by " + str(divisible))
        else:
            print("The number is not divisible by " + str(divisible))        
    else:
        multiple=randint(3,7)
        if number%multiple == 0:
            print("The number is a multiple of " + str(multiple))
        else:
            print("The number is not multiple of " + str(multiple))  
        
```

```python
number= randint(1,100)
user_guess= int(input("Introduce a number between 1 and 100: "))
user_guess=check_user_guess(user_guess)
right_guess= compare_number(number, user_guess)

while (right_guess == False):
    print("Woops! Looks like you guessed the wrong number! Here you have a hint.")
    generate_clue(number)
    user_guess=int(input("Try with anoter number: "))
    user_guess=check_user_guess(user_guess)
    right_guess= compare_number(number, user_guess)

print ("Congrats! You guessed the number!")
```
